```python
from pdfminer.pdfinterp import PDFResourceManager, PDFPageInterpreter
from pdfminer.converter import TextConverter
from pdfminer.layout import LAParams
from pdfminer.pdfpage import PDFPage
from io import StringIO
##from io import BytesIO
import codecs
import pandas as pd
```

```python
# Lendo os arquivos texto
from os import listdir
from os.path import isfile, join
mypath="C:/AI/Krypton-01/texto"
onlyfiles1 = [f for f in listdir(mypath) if isfile(join(mypath, f))]
```

```python
documentos1 = {'tipo'  : [""],
               'nome'  : [""],
               'array1': [""]
        }
df1 = pd.DataFrame(documentos1, columns= ['tipo','nome','array1'])
```

```python
contnota=0
contador1 = 0

for file1 in onlyfiles1:
    final1 = (file1[len(file1)-3:])
    nome1  = (file1[:len(file1)-4])
    if final1=='txt':
        ttipo1="NOT IDENT"
        print(nome1)
        ##my_file = open(r"C:/AI/krypton-01/texto/"+nome1+".txt","r+")
        my_file = open("C:/AI/krypton-01/texto/"+nome1+".txt", 'r', encoding="utf8").read()
        ##print(my_file.read())
        print(len(my_file))
        ### Identificacao se é nota e o numero 
        palavra1=""
        for i in range(len(my_file)):
            letra=my_file[i]
            palavra1=palavra1+letra
            if len(palavra1)>6:
                palavra1=palavra1[1:]
            if palavra1.upper() == "FISCAL":
                contnota=contnota+1
                print("achei uma nota")
                print(palavra1)
                ttipo1 = "NOTA"
 
        #################################
        ## Identificação recibo
        palavra12=""
        for i2 in range(len(my_file)):
            letra=my_file[i2]
            palavra12=palavra12+letra
            if len(palavra12)>6:
                palavra12=palavra12[1:]
                if palavra12.upper() == "RECIBO":
                    print("achei um recibo")
                    print(palavra12)
                    ttipo1 = "RECIBO"
                    
        ## Fim Identificação doc
        #Identificacao valor liquido
        palavra13=""
        for i3 in range(len(my_file)):
            letra=my_file[i3]
            palavra13=palavra13+letra
            if len(palavra13)>5:
                palavra13=palavra13[1:]
                if palavra13.upper() == "BANCO":
                    print("achei um DOC")
                    print(palavra13)
                    ttipo1="DOC"
    df1 = df1.append({'tipo': ttipo1, 'nome':nome1}, ignore_index=True)
```
