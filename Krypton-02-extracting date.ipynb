```python
def is_number(s):
    try:
        float(s)
        return True
    except ValueError:
        return False
```

```python
from nltk.tokenize import word_tokenize
df1.dropna(subset=['body'],inplace=True)
df1.reset_index(inplace = True, drop = True)
df1['value']=""
len5 = len(df1)
for i in range(len5):
    maior=0
    print(i)
    #zz= clean1(str(df3["body"][i]))
    text1 = str(df1["body"][i])
    text1 =  text1.encode('utf-8-sig')
    words = nltk.word_tokenize(df1["body"][i])
    body1=""
    count=0
    for i1 in words:
        lenx=len(i1)
        wordy=''
        ver=0
        for i2 in range(lenx):
            letter=i1[i2]
            if letter==".":
               letter=""
            if letter==",":
                ver=1
                letter="."
            wordy=wordy+letter
        nx=is_number(wordy)
        if nx==True and ver==1:
            #print('é numerico',wordy)
            if float(wordy)>maior:
                maior=float(wordy)
            
    count=count+1
    df1['value'][i]=maior
                
```
