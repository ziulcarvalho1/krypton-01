```python
def is_number(s):
    try:
        float(s)
        return True
    except ValueError:
        return False
```

```python
import string
def check(value):  
    for letter in value:  
             
        # If anything other than digit  
        # letter is present, then return  
        # False, else return True  
        if letter not in string.digits:  
            return False
    return True
```
