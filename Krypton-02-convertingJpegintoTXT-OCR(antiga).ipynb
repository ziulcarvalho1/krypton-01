```python
from wand.color import Color
from wand.image import Image as Img
from PIL import Image 
import pytesseract 
import sys 
from pdf2image import convert_from_path 
from pdf2image import convert_from_path, convert_from_bytes
from pdf2image.exceptions import (
    PDFInfoNotInstalledError,
    PDFPageCountError,
    PDFSyntaxError
)
import os 
import cv2
import numpy as np
```

```python
from os import listdir
from os.path import isfile, join
mypath="C:/AI/Krypton-01/jpeg2"
onlyfiles = [f for f in listdir(mypath) if isfile(join(mypath, f))]
```

```python
#Usando magik
import tempfile
import re
counter=0
for file in onlyfiles:
    final = (file[len(file)-3:])
    nome  = (file[:len(file)-4])
    #print(final)
    if final=='png':
        counter=counter+1
        print(counter,nome)
        if counter>106:
            #origem = "C:/AI/krypton-01/jpeg2/"+nome+".jpeg"
            origem1 ="C:/AI/krypton-01/jpeg2/"+nome+".png"
            destino= "C:/AI/krypton-01/texto2/"+nome+".txt"
            im = Image.open(origem1)
            #im.save(origem1, dpi=(96, 96))
            image = cv2.imread(origem1)
            image = cv2.resize(image, None, fx=2, fy=2, interpolation=cv2.INTER_CUBIC)
            retval, threshold = cv2.threshold(image,127,255,cv2.THRESH_BINARY)
            pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files (x86)\Tesseract-OCR\tesseract.exe'
            text = pytesseract.image_to_string(threshold,lang='por')
            with open(destino, 'w',5 ,'utf-8') as text_file:
                text_file.write(text)
```
