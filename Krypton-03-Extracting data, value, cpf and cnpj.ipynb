```python
from dateutil.parser import parse
import re
df1.dropna(subset=['body1'],inplace=True)
df1.reset_index(inplace = True, drop = True)
df1['date']=""

  
def most_frequent(List):
  new_list = []
  for date in List:
    if "-" in date:
      day, month, year = map(int, date.split("-"))
    else:  
      day, month, year = map(int, date.split("/"))
    if 1 <= day <= 31 and 1 <= month <= 12 and 2000 <= year <= 2021:
      new_list.append(date)
    if len(new_list) > 0:
      counter = 0
      num = new_list[0]
      for i in new_list:
        curr_frequency = new_list.count(i) 
        if (curr_frequency > counter): 
          counter = curr_frequency 
          num = i
      return num 
    if len(new_list) == 0:
      return "no date"


len1 = len(df1)
print(len1)
for i in range(len1):
    content = df1['body1'][i]
    # The regex pattern that we created
    pattern = "\d{2}[/-]\d{2}[/-]\d{4}"
    dates = re.findall(pattern, content)
    date = most_frequent(dates)
    print(df1['name'][i])
    df1['date'][i]=date
    print(df1['date'][i])
```

```python
def is_number(s):
    try:
        float(s)
        return True
    except ValueError:
        return False
```

```python
from nltk.tokenize import word_tokenize
df1.dropna(subset=['body1'],inplace=True)
df1.reset_index(inplace = True, drop = True)
df1['value']=""
len5 = len(df1)
for i in range(len5):
    maior=0
    print(i)
    #zz= clean1(str(df3["body"][i]))
    text1 = str(df1["body1"][i])
    text1 =  text1.encode('utf-8-sig')
    words = nltk.word_tokenize(df1["body1"][i])
    body1=""
    count=0
    for i1 in words:
        lenx=len(i1)
        wordy=''
        ver=0
        for i2 in range(lenx):
            letter=i1[i2]
            if letter=='$':
                letter=""
            if letter=="R":
                letter=""
            if letter==".":
                letter=""
            if letter==",":
                ver=1
                letter="."
            wordy=wordy+letter
        nx=is_number(wordy)
        if nx==True and ver==1:
            #print('é numerico',wordy)
            if float(wordy)>maior:
                maior=float(wordy)
            
    count=count+1
    df1['value'][i]=maior
```

```python
import string
def check(value):  
    for letter in value:  
             
        # If anything other than digit  
        # letter is present, then return  
        # False, else return True  
        if letter not in string.digits:  
            return False
    return True
```
